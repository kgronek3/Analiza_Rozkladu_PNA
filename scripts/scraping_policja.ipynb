```python
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
import time
```

```python
url = 'https://policja.pl/pol/form/1,informacja-dzienna.html?page=' + str(0)
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
table = soup.find('table', {'class': 'table-listing table-striped margin_b20'})
headers = []

for i in table.find_all('th'):
    title = i.text
    headers.append(title)

df = pd.DataFrame(columns = headers)
df
```

```python
for a in range(42):
    for i in range(336)[a*8:a*8+8]:
        url = 'https://policja.pl/pol/form/1,informacja-dzienna.html?page=' + str(i)
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'lxml')
        table = soup.find('table', {'class': 'table-listing table-striped margin_b20'});

        for row in table.find_all('tr')[1:]:
            data = row.find_all('td')
            row_data = [td.text.strip() for td in data]
            length = len(df)
            df.loc[length] = row_data
    time.sleep(10)
            
```

```python
df.to_csv("../data/policja.csv", sep = ';', encoding = 'utf-8', index = None, header = True)
```
